```python
import tweepy
import pickle

from twitter_keys import consumer_key, consumer_secret, access_token, access_secret
auth=tweepy.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_secret)

api=tweepy.API(auth)
pawanKalyan_tweets = api.user_timeline('@PawanKalyan')

with open('pawanKalyan_tweets.pkl','wb') as f:
    pickle.dump(pawanKalyan_tweets,f)
```

```python
MaheshGinugu_tweets = api.user_timeline('@MaheshGinugu')

with open('MaheshGinugu_tweets.pkl','wb') as f:
    pickle.dump(MaheshGinugu_tweets,f)
```

```python
def lexical_diversity(text):
    tokens=text.lower().split()
    return len(set(tokens))/len(tokens)

pawanKalyan_diversities=[]
for tweet in pawanKalyan_tweets:
    pawanKalyan_diversities.append(lexical_diversity(tweet.text))

#print(len(pawanKalyan_tweets))
# print(pawanKalyan_diversities)
```

```python
PSPKFan2You_diversities=[]
for tweet in PSPKFan2You_tweets:
    PSPKFan2You_diversities.append(lexical_diversity(tweet.text))

# print(len(PSPKFan2You_tweets))
# print(len(PSPKFan2You_diversities))
# print(PSPKFan2You_diversities)
```

```python
MaheshGinugu_diversities=[]
for tweet in MaheshGinugu_tweets:
    MaheshGinugu_diversities.append(lexical_diversity(tweet.text))

# print(len(MaheshGinugu_tweets))
# print(len(MaheshGinugu_diversities))
# print(MaheshGinugu_diversities)
```

```python
pawanKalyan_total_diversity = 0.0
all_pawanKalyan_tweets = ""


for i in range(len(pawanKalyan_tweets)):
    all_pawanKalyan_tweets += pawanKalyan_tweets[i].text
    all_pawanKalyan_tweets += " "
    
pawanKalyan_total_diversity = lexical_diversity(all_pawanKalyan_tweets)
# print("Total lexical diversity of Pawan Kalyan's tweets:", pawanKalyan_total_diversity)
```

```python
PSPKFan2You_total_diversity = 0.0
all_PSPKFan2You_tweets = ""


for i in range(len(PSPKFan2You_tweets)):
    all_PSPKFan2You_tweets += PSPKFan2You_tweets[i].text
    all_PSPKFan2You_tweets += " "
    
PSPKFan2You_total_diversity = lexical_diversity(all_PSPKFan2You_tweets)
# print("Total lexical diversity of PSPKFan2You's tweets:", PSPKFan2You_total_diversity)
```

```python
KapilVarala_total_diversity = 0.0
all_KapilVarala_tweets = ""


for i in range(len(Kapilvarala_tweets)):
    all_KapilVarala_tweets += Kapilvarala_tweets[i].text
    all_KapilVarala_tweets += " "
    
KapilVarala_total_diversity = lexical_diversity(all_KapilVarala_tweets)
# print("Total lexical diversity of Kapil Varala's tweets:", KapilVarala_total_diversity)
```

```python
MaheshGinugu_total_diversity = 0.0
all_MaheshGinugu_tweets = ""


for i in range(len(MaheshGinugu_tweets)):
    all_MaheshGinugu_tweets += MaheshGinugu_tweets[i].text
    all_MaheshGinugu_tweets += " "
    
MaheshGinugu_total_diversity = lexical_diversity(all_MaheshGinugu_tweets)
# print("Total lexical diversity of Mahesh Ginugu's tweets:", MaheshGinugu_total_diversity)
```

```python
import numpy as np
import matplotlib.pyplot as plt

font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 10,
        }

people = ("Pawan Kalyan", "PSPKFan2You", "Kapil Varala", "Mahesh Ginugu")
y_pos = np.arange(len(people))
Lexical_diversity_value = [pawanKalyan_total_diversity, PSPKFan2You_total_diversity, KapilVarala_total_diversity, MaheshGinugu_total_diversity]

plt.bar(y_pos, Lexical_diversity_value, width=0.5, align='center', alpha=0.75, color=['red', 'blue', 'green', 'yellow'], edgecolor='black')
for a,b in zip(y_pos, Lexical_diversity_value):
    plt.text(a, b, str(b), fontdict=font)
plt.xticks(y_pos, people)
plt.ylabel('Lexical_Diversity_Values')
plt.title('Lexical Diversity')

import matplotlib.style as style
style.use('default')
style.use('seaborn-talk')

plt.show()
```

```python
import nltk
nltk.download('vader_lexicon')


from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

polarity_score_pawanKalyan = analyzer.polarity_scores(all_pawanKalyan_tweets)
polarity_score_PSPKFan2You = analyzer.polarity_scores(all_PSPKFan2You_tweets)
polarity_score_KapilVarala = analyzer.polarity_scores(all_KapilVarala_tweets)
polarity_score_MaheshGinugu = analyzer.polarity_scores(all_MaheshGinugu_tweets)

# print("Sentiment Analysis for Pawan Kalyan's tweets:", polarity_score_pawanKalyan)
# print("Sentiment Analysis for PSPKFan2You's tweets:", polarity_score_PSPKFan2You)
# print("Sentiment Analysis for Kapil Varala's tweets:", polarity_score_KapilVarala)
# print("Sentiment Analysis for Mahesh Ginugu's tweets:", polarity_score_MaheshGinugu)
```

```python
people = ("Pawan Kalyan", "PSPKFan2You", "Kapil Varala", "Mahesh Ginugu")
y_pos = np.arange(len(people))

font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 10,
        }

polarity_score_pawanKalyan_list = list(polarity_score_pawanKalyan.values())
polarity_score_PSPKFan2You_list = list(polarity_score_PSPKFan2You.values())
polarity_score_KapilVarala_list = list(polarity_score_KapilVarala.values())
polarity_score_MaheshGinugu_list = list(polarity_score_MaheshGinugu.values())

sentiment_analysis_compound_value = [polarity_score_pawanKalyan_list[3], polarity_score_PSPKFan2You_list[3], polarity_score_KapilVarala_list[3], polarity_score_MaheshGinugu_list[3]]

plt.bar(y_pos, sentiment_analysis_compound_value, width=0.5, align='center', alpha=0.75, color=['red', 'blue', 'green', 'yellow'], edgecolor='black')

for a,b in zip(y_pos, sentiment_analysis_compound_value):
    plt.text(a, b, str(b), fontdict=font)

plt.xticks(y_pos, people)
plt.ylabel('Sentiment_Analysis_Value (Compound)')
plt.title('Sentiment Analysis (Compound)')

import matplotlib.style as style
style.use('default')
style.use('seaborn-talk')

plt.show()
```
